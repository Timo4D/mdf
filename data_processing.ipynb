```python
import pandas as pd
```

```python
register = pd.read_csv('data/register.csv')
register.head()
```

```python
print(register.columns)
```

```python
register = register.drop(columns=['Unnamed: 9',
       'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12'])
register.head()
```

```python
# Summary description of the 'group' column in the existing `register` DataFrame
s = register['group']

print("dtype:", s.dtype)
print("non-null:", s.count(), "/", len(s))
print("nulls:", s.isnull().sum())

print("\nValue counts:")
print(s.value_counts(dropna=False))

print("\nRelative frequencies (%):")
print((s.value_counts(normalize=True, dropna=False) * 100).round(2))

print("\nUnique values:")
print(s.unique())

print("\nDescribe():")
print(s.describe())
```

# Because of imbalance I will use Faller(FHs, FHm) vs Non Faller(HA+NF) as the target.

```python
# create a column 'faller' in the register DataFrame: 1 for FHs/FHm, 0 for HA/NF
register['faller'] = register['group'].isin(['FHs', 'FHm']).astype(int)

# quick check
register[['part_id', 'group', 'faller']].head()
```

```python
# replace 'm'/'f' with 0/1 (case-insensitive) and make column integer (nullable)
register['sex'] = register['sex'].astype(str).str.strip().str.lower().map({'m': 0, 'f': 1})
register['sex'] = register['sex'].astype('Int64')

# quick check
register[['part_id', 'sex']].head()
```

```python
register['part_id'] = register['part_id'].str.replace('SPPB', '', regex=False)

# quick check
register[['part_id', 'group', 'faller']].head()
```

```python
register.to_csv('data/register_processed.csv', index=False)
```
